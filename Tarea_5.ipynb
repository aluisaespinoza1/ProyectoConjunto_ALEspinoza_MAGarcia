Código de solución Ana Luisa

La función $g(x)$ se define como: $$g(x) = \frac{t(x)}{c}$$

El valor de $c$ es una constante resultante de la siguiente expresión: $$c=\int_{-\infty}^{\infty}t(x) dx $$

En este caso, la integral de la función $t(x) = a \sin(\pi x)$ tiene límites definidos donde $a = 0$ y $b = 1$ por lo que:
 $$c=\int_{0}^{1} a \sin(\pi x)  dx = \frac{2a}{\pi} $$
 
Entonces, la función $g(x)$ es: $$g(x) = \frac{a \sin(\pi x)}{\frac{2a}{\pi}} = \frac{a \sin(\pi x) \pi}{2a} = \frac{\pi \sin(\pi x)}{2}$$

### Ejercicio 6: 
6. Comparar el porcentaje de puntos de aceptación cuando se usa $t(x)$ constante y $t(x)$ un pulso senoidal. Concluir

Código de solución Ana Luisa

```python
def porcentaje_A_R(N:'número de puntos',
    f:'función objetivo a generar',
    t:'función que mayora a f'):
    
    R2 = np.random.rand(N)
    R1 = np.random.rand(N)
    f_x = f(R1)
    t_x = t(R1)
    condition = R2*t_x <=f_x
    acept = [R1[i] for i in range(len(R1)) if condition[i]]
    P = len(acept)/N
    return P
```

```python
N = 500
f = lambda x: 30 * (x**2 -2 * x**3 + x**4)
a = f(optimize.fmin(lambda x:-f(x), 0, disp=False))

t_sen = lambda x: a*np.sin(np.pi*x)
t_c = lambda x: a * np.ones(len(x))

p_senoidal = porcentaje_A_R(N,f,t_sen)
constante = porcentaje_A_R(N,f,t_c)
```

```python
print(f"El porcentaje de puntos de aceptación usando un pulso senoidal como t(x) es de: {p_senoidal*100}%")
print(f"El porcentaje de puntos de aceptación usando una constante como t(x) es de: {constante*100}%")
```

### Ejercicio 8: 
8. Demostrar la validez de la siguiente ecuación
$$p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p} p_i$$

Código de solución Ana Luisa

Tomando en cuenta las siguientes propiedades de los factoriales:



Partiendo de la expresión:
$$ p_i=P(X=i)=\frac{n!}{i!(n-i)!}p^i(1-p)^{n-i} $$ 

Sustituyendo en la expresión $i$ por $i+1$ la ecuación queda de la siguiente forma:

$$ p_{i+1} = \frac{n!}{(i+1)!(n-(i+1))!}p^{i+1}(1-p)^{n-(i+1)} $$

Dividiendo este término entre el anterior:

$$ p_{i+1} = \frac{\frac{n!}{(i+1)!(n-(i+1))!}p^{i+1}(1-p)^{n-(i+1)}}{\frac{n!}{i!(n-i)!}p^i(1-p)^{n-i}} = \frac{(n-i)p}{(1+i)(1-p)}p_i $$

Entonces la ecuación queda:
$$ p_{i+1} = \frac{n-i}{i+1}\frac{p}{1-p} p_i$$

9. Seguir un procedimiento similar al mostrado cuando se generó una distribución binomial, pero en esta caso genere un código que genere variables aletorias Poisson cuya función de distribución de probabilidad esta dada por:
$$P(k,\lambda)=\frac{e^{-\lambda}(\lambda)^k}{k!}$$
    

Demuestre matemáticamente que 
 $$P(k+1)=\frac{\lambda}{k+1}P(k)$$
 y a partir de esta relación genere variables aletorias que distribuyen poisson usando el método de la transformada inversa.

Código de solución Ana Luisa

Partiendo de la ecuación $P(k,\lambda)=\frac{e^{-\lambda}(\lambda)^k}{k!}$, suponiendo que $\lambda$ es constante tenemos:

$$ P(k)=\frac{e^{-\lambda}(\lambda)^k}{k!} $$

Sustituyendo $k$ por $k+1$ la ecuación queda: 

$$P(k+1)=\frac{e^{-\lambda}(\lambda)^{k+1}}{(k+1)!}$$
 
Dividiendo entre el término anterior:

$$ P(k+1)= \frac{\frac{e^{-\lambda}(\lambda)^{k+1}}{(k+1)!}}{\frac{e^{-\lambda}(\lambda)^k}{k!}} $$
$$ P(k+1)= \frac{e^{-\lambda}(\lambda)^{k}\lambda}{(k+1)k!} = \frac{e^{-\lambda}(\lambda)^k}{k!}\frac{\lambda}{k+1}$$

En la ecuación anterior se puede observar que se encuentra el término de $P(K)$, entonces sustituyendo:

 $$P(k+1)=\frac{\lambda}{k+1}P(k)$$

```python
def Poisson(k,l):
    global pr
    pr = np.zeros(k)
    pr[0] = np.exp(-l)
    def Pr(i):
        global pr
        pr[i+1]=l/(i+1)*pr[i]
        
    # Lleno el vector Pr usando compresión de listas
    [Pr(i) for i in range(k-1)]
    return np.cumsum(pr)

```

```python
N = 10000
k, lam = 17, 6
poisson = Poisson(k,lam) #prob acumulada distr Poisson


u = np.random.rand(N)

distribucion = list(map(lambda j:sum(1 for i in poisson if i<u[j]),range(N)))


plt.figure(figsize=[10,4])
y, x = np.histogram(distribucion, bins=len(set(distribucion)), density=True) 
plt.bar(x[1:], y, label='Distribución de Poisson')
plt.legend()
plt.show()
    
```
